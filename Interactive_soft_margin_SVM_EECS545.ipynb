```
from sklearn.datasets import make_blobs
from sklearn import svm
import numpy as np


n_samples = 200
outliers_fraction = 0.01
n_outliers = int(outliers_fraction * n_samples)
n_inliers = n_samples - n_outliers
X, y = make_blobs(n_samples=n_inliers, centers=[(-3, 2), (0,0)], random_state=0, cluster_std=0.50)
y[y == 0] = -1
tmp = np.ones(len(X))
y = tmp * y

#Ploting the dataset without outliers
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.grid()
plt.title("Dataset without outliers")
plt.show()

#Adding outliers..as required
x_outlier=[(-1.85,0.5),(-1.5,1.7),(-1.20,2),(-1.3,1)]
X = np.concatenate([X, x_outlier], axis=0)
#Labelling the outliers...as required
y=  np.concatenate([y,[0,0,1,1]],axis=0)
y[y == 0] = -1
tmp = np.ones(len(X))
y = tmp * y                 
#Ploting the dataset with outliers
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.grid()
plt.title("Dataset with outliers")
plt.show()

```

```
def plot_portfolio(C=100):
    
    h = 0.01 # step size in the mesh

    # we create an instance of SVM and fit out data. We do not scale our
    # data since we want to plot the support vectors
    #C = 100 # SVM regularization parameter
    svc = svm.SVC(kernel='linear', C=C).fit(X, y)
   

    # create a mesh to plot in
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max,h), np.arange(y_min, y_max,h))
    print(xx.shape)
    

    Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
    
    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    #plt.contour(xx, yy, Z, alpha=0.8)

    plt.contourf(xx, yy, Z, alpha=0.3)

    #Ploting the dataset with outliers
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.grid()

  
```

```
%%time
plot_portfolio(100)   #Time required for function execution
```
